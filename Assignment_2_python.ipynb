**Q4**. Create a function heart_rate_stats(max_bpm, avg_bpm, resting_bpm) that returns:

● Heart Rate Reserve = max_bpm - resting_bpm

● Average Difference = avg_bpm - resting_bpm

Print both returned values separately.

```python
def heartratestats(maxbpm, avgbpm, restingbpm):
    hrr = maxbpm - restingbpm
    avg_diff = avgbpm - restingbpm
    return hrr, avg_diff # takes 3 parameters and returns 2 values

hrr, avg_diff = heartratestats(190, 140, 70)
print(hrr)
print(avg_diff)
```

**Q6**. Create two functions:

**Q12**. Write a function daily_burns(calories_list) that prints burn for each day and returns total.

**Q17**. Write a recursive function sum_n(n) to return sum of first n natural numbers.
Call sum_n(100).

```python
def sum_n(n):
    if n == 1:
        return 1
    return n + sum_n(n - 1)
#recursion-> calls the funcn inside another funcn
print(sum_n(100))
```

**Q18**. Write a recursive function factorial(n) to return n!
Call factorial(6).

```python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
# uses n-1! to cal n!
print(factorial(6))
```

**Q23**. Write a function to visualize Calories_Burned vs Session_Duration for each Workout_Type using Matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
def visualize():
  a=pd.read_csv("Final_data.csv")#reading the data
  a=a.groupby("Workout_Type")#grouping the data to visualize
  for i,j in a:
    plt.plot(j["Session_Duration (hours)"],j["Calories_Burned"],label=i)#plotting each workout_type graph
  plt.legend=("Workout_Type")
  plt.xlabel("Session_Duration")#labeling the graph
  plt.ylabel("Calories_Burned")
  plt.show()
visualize()
```

**Q25**.Write a function that creates and saves any Matplotlib plot as a PNG file automatically

```python
import pandas as pd
import matplotlib.pyplot as plt
def save_plot():
  a=pd.read_csv("Final_data.csv")
  b=a.head()
  plt.bar(b["Weight (kg)"],b["Height (m)"])#plotting the graph
  plt.title('Weight vs height')#labelling the graph
  plt.xlabel('weight')
  plt.ylabel('Height')
  plt.grid(True)
  plt.show()
  plt.savefig("plot.png")#method for saving the plot as png file
  print("plot saved as plot.png")
save_plot()
```
