# **1. 5 applcns of xor**
# **2. is vs ==**



1.Bitwise Swapping (without a temporary variable):

XOR can swap two numbers without using extra space.




2.Finding the Odd Occurring Element in an Array:

If every element appears twice except one, XOR of all elements gives the unique one (because x ^ x = 0 and x ^ 0 = x).

3.Checksums / Error Detection:

Used in simple error detection (parity bits, LRC, CRC) because XOR highlights bit changes.

4.Encryption / Decryption:

XOR with a key is used in many stream ciphers and lightweight encryption (since cipher = data ^ key and data = cipher ^ key).

5.Detecting Bit Differences:



XOR shows which bits differ between two numbers (1 where bits differ, 0 where same). Useful in Hamming distance and image processing.

In Python,

==

Checks for equality of values.

It compares whether the contents of two objects are the same.

Example:

a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)   # True (contents are equal)
---
is

Checks for identity (whether both variables point to the same object in memory).

It compares object references, not values.

Example:

a = [1, 2, 3]
b = [1, 2, 3]

print(a is b)   # False (different objects in memory)
print(a == b)   # True  (contents are equal)

c = a
print(a is c)   # True (both refer to the same object)



---

Use == → when you care about values.

Use is → when you care about object identity (whether they are the same object).

arbitrary args - (*args)

any no of arguments can be used

*args mai jo input jata hai wo tuple mai jata hai


```python
def my_fam(*args):
  print("youngest child is "+ args[1])
  print(type(args))
  print("first argument:"+ args[0])
my_fam("a","b","c")
```

```python
def sec_d(greeting,*names):
  for name in names:
   print(greeting,name)
sec_d("hi","amit","swati","shiva")
```

```python
def total(*args):
  total=0
  for i in args:
    total=total+i
  return total
print(total(25,26,27,28,29))
```

```python
def max_val(*args):
  max=args[0]
  for i in args:
    if i>max:
      max=i
  return max
print(max_val(1,2,3,4,5,6))
```

****kwargs**
key word args

```python
def fam(**kwargs):
  print("type:",type(kwargs))
  print("name:", kwargs["name"])
  print("age:",kwargs["age"])
  print(kwargs)
fam(name="swati",age=20,dept="ece")
#
```

```python
def my_fun(title,*args,**kwargs):
  print("title:",title)
  print("positional:",args)
  print("keywords:",kwargs)
my_fun("user info","nitika","swati","xyz",age="29",city="vskp")
```
