# 1. Write a program to find the first non-repeating character in a string using a for loop and else.


```python
 s = "swiss"

for i in range(len(s)):
    if s.count(s[i]) == 1:
        print("First non-repeating character:", s[i])
        break
else:
    print("No non-repeating character found.")
```

# Given a number n, write a program to check if it is prime using a for-else loop.



```python
n = 150
count = 0

for i in range(1, n + 1):  # Check from 1 to n inclusive
    if n % i == 0:
        count += 1

if count == 2:
    print("prime")
else:
    print("not prime")

```

#  Implement a program to find the sum of digits of a number using a while loop and continue statement to skip zeros.




```python
# n=350
# sum=0
# rem=0
# while n>0:
#   rem=n%10
#   n=int(n/10)
#   sum+=rem
# print(sum)

n = int(input("Enter a number: "))
sum_digits = 0
while n > 0:
    digit = n % 10
    n = n // 10

    if digit == 0:
        continue  # Skip zero digits

    sum_digits += digit

print("Sum of non-zero digits:", sum_digits)

```

#  Implement a program that checks if a string is a palindrome using for-else.


```python
s = input("Enter a string: ")
length = len(s)

for i in range(length // 2):
    if s[i] != s[length - 1 - i]:
        print("Not a palindrome")
        break
else:
    print("Palindrome")
```

#   Write a program to count vowels in a string using for loop. Use continue to skip consonants.

```python
s = input("enter a string: ")
length = len(s)
count = 0
for i in range(length):
  if s[i] in ('a', 'e', 'i', 'o', 'u'):
    count += 1
  else:
    continue
print(count)
```

# Write a program to count the number of words in a sentence using for loop.


```python
sentence = input("Enter a sentence: ")
word_count = 0
in_word = False

for char in sentence:
    if char.isspace():
        in_word = False
    elif not in_word:
        word_count += 1
        in_word = True

print("Number of words:", word_count)
```

# Write a program to check if a number is an Armstrong number using for-else.

```python
n = int(input("Enter a number: "))
s = str(n)
l = len(s)
total = 0

for d in s:
    total += int(d) ** l
else:
    if total == n:
        print("Armstrong number")
    else:
        print("Not an Armstrong number")
# validate i/p

```
